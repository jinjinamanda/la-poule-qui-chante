```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn import preprocessing
import matplotlib.pyplot as plt
dfDisponibiliteAlimentaire = pd.read_csv('DAN-P9-data/DisponibiliteAlimentaire_2017.csv', keep_default_na=False)
dfPopulation = pd.read_csv('DAN-P9-data/Population_2000_2018.csv', keep_default_na=False)

import warnings
warnings. simplefilter(action='ignore', category=Warning)
```

```python
dfDisponibiliteAlimentaire.head()
```

```python
dfImportationProduction = dfDisponibiliteAlimentaire.loc[dfDisponibiliteAlimentaire["Élément"].isin(["Importations - Quantité", "Production"])]
dfImportationProduction.head()
```

```python
dfImportationProductionViandevolaille = dfImportationProduction.loc[dfImportationProduction["Produit"] == "Viande de Volailles"]
dfImportationProductionViandevolaille.head()
```

```python
dfpivot_tablevolaille = dfImportationProductionViandevolaille.pivot_table(values="Valeur", index=["Zone","Produit"], columns="Élément")
dfpivot_tablevolaille.head()
```

```python
dfPopulation.head()
```

```python
dfPopulation.describe()
```

```python
dfPopulationsort= dfPopulation.sort_values(["Zone","Année"], ascending=True)
dfPopulationsort.head(10)
```

```python
dfPopulationsort["Évolution population"] = dfPopulationsort["Valeur"].diff(1)/dfPopulationsort["Valeur"].shift(1)

dfPopulationsort["Valeur"] = dfPopulationsort["Valeur"].replace(to_replace=2000,value=np.nan)

dfPopulation20012017 = dfPopulationsort.loc[(dfPopulationsort["Année"] >= 2001) & (dfPopulationsort["Année"] <= 2017)]

dfmoyenne_evolution_population = dfPopulation20012017["Évolution population"].mean()

print(dfmoyenne_evolution_population)
```

```python
dfPopulation2017 = dfPopulation20012017.loc[(dfPopulation20012017["Année"] == 2017)]
dfPopulation2017.head()
```

```python
dfPopulation2017['Valeur']= dfPopulation2017['Valeur']*1000
dfmergedviandevolaille = pd.merge(dfpivot_tablevolaille, dfPopulation2017[['Zone','Valeur','Évolution population']], on="Zone")
dfmergedviandevolaille.head()
```

```python
# Créer un ensemble de données
data = np.random.randn(1000)

# Créer un boxplot
plt.boxplot(dfmergedviandevolaille)

# Identifier les outliers
outliers = [x for x in data if x < data.quantile(0.25) - 1.5 * data.iqr() or x > data.quantile(0.75) + 1.5 * data.iqr()]

print(outliers)
```

```python
plt.plot(dfmergedviandevolaille.Zone,dfmergedviandevolaille.Valeur, "o")
plt.plot(np.arange(100),[a*x+b for x in np.arange(100)])
plt.xlabel("Zone")
plt.ylabel("Valeur")
plt.show()
```

```python
dfmergedviandevolaille['scalerPop']=preprocessing.RobustScaler().fit_transform(dfmergedviandevolaille[['Valeur']])
dfmergedviandevolaille.head()
```

```python
plt.plot(dfmergedviandevolaille.Zone,dfmergedviandevolaille.scalerPop, "o")
plt.plot(np.arange(100),[a*x+b for x in np.arange(100)])
plt.xlabel("Zone")
plt.ylabel("scalerPop")
plt.show()
```

```python
# Importez le modèle
from sklearn.neighbors import LocalOutlierFactor

# Créez le modèle
model = LocalOutlierFactor(n_neighbors=5)

# Supprimez les valeurs NaN
dfmergedviandevolaille = dfmergedviandevolaille.dropna()

# Calculez les scores d'outliers
scores = model.fit_predict(dfmergedviandevolaille[['Importations - Quantité', 'Valeur', 'Production']])

# Identifiez les outliers
outliers = dfmergedviandevolaille[scores == -1]

# Comptez le nombre d'outliers
n_outliers = len(outliers)

# Imprimez les outliers
print(outliers)
```
